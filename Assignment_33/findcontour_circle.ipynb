```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("input/6.png"   )
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow( img )
plt.show()
```

```python
_ , thresh = cv2.threshold(img , 50 , 255 , cv2.THRESH_BINARY)
plt.imshow(thresh , cmap="gray")
plt.show()
```

```python
print(img.shape[0] , img.shape[1])
height = img.shape[0] 
width  = img.shape[1]  

for i in range(height-5):
     for j in range(width-5):
          if thresh[i,j].all() != thresh[i,j+1].all() :
                    #thresh[i:i+1 , j-1:j+1] = (255,0,0)
                    thresh[i:i+1 , j-6:j+1] = (255,0,0)
                    
for j in range(width-5):
     for i in range(height-5):
          if thresh[i,j].all() != thresh[i+1,j].all() :
               #thresh[i-1:i+1 , j:j+1] = (255,0,0)  
               thresh[i-6:i+1 , j:j+1] = (255,0,0)  

plt.imshow(thresh , cmap="gray")
plt.show()
```
