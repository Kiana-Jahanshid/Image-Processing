```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("input/c8.jpg" , cv2.IMREAD_GRAYSCALE)
plt.imshow( img , cmap="gray")
plt.axis("off")
plt.show()
```

```python
_ , thresh = cv2.threshold(img , 100 , 255 , cv2.THRESH_BINARY)
plt.imshow( thresh , cmap="gray")
plt.axis("off")
plt.show()
```

```python

# noise reduction , ba blur / median 
# threshold zadan
# filter gozari rooye size 
```

```python
contours , hierarchy = cv2.findContours(thresh , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_SIMPLE) # ax midim , liste contour ha ro mide 
print(len(contours))
print(hierarchy)
# agar -1 bashe yani contour external hast 
# oon khati ke index akharesh 1 hastan seke han  
coins = img.copy()
count = 0
for contour in contours :
    if cv2.contourArea(contour) > 100 :
        x , y , w , h = cv2.boundingRect(contour)
        cv2.rectangle(coins , (x,y) , (x+w , y+h) , (127 , 127 , 127) , 3)
        count+=1
print(count)
plt.axis("off")  
plt.imshow(coins , cmap="gray") 
plt.show()    
```
