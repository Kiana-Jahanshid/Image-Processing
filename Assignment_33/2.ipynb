```python
img = cv2.imread("rose.jpg" , cv2.IMREAD_GRAYSCALE)
img1 = cv2.imread("w.jpg"   , cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("spider.png" , cv2.IMREAD_GRAYSCALE)
img4 = cv2.imread("p.png" , cv2.IMREAD_GRAYSCALE)
img3 = np.zeros([100,200] , dtype=np.uint8)

plt.imshow( img4 , cmap="gray")
#plt.axis("off")
plt.show()
```

```python
_ , thresh = cv2.threshold(img4 , 127 , 255 , cv2.THRESH_BINARY)
plt.imshow( thresh , cmap="gray")
plt.show()
```

```python
invert = 255 - thresh
plt.imshow(invert , cmap="gray")
plt.show()
```

```python
# image , mode , method , 
# mode : """EXTERNAL contour""" : yani contours haye birooni ro mide ba ma 
#      : """LIST"""              : all contours ro barmigardoone
#      : """TREE"""              : migam kodoom contour , dakhele kodoom contour hast . hame contour ha ro behemoon mide ,  nesbate pedar farzandi shoono h am be ma miide 


# method : 
# vaghti contour return mikone chia return mikone ?
# CHAIN_APPROX_NONE :   STORES ALL THE CONTOURS POINTS .  tamame pixel haye rec ro barmigardoone --------> PRINT KON CONTOURS RO === TAK TAK PIXEL H
# CHAIN_APPROX_SIMPLE :  noghate atraf ya gooshe rectangle ro mide  , faghat 4 NOGHTE gooshe ro mide 
# CHAIN_APPROX_TC89_L1 : 
# CHAIN_APPROX_TC89_KCOS : 


contours  , _ =  cv2.findContours(invert , cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE)
res = img4.copy()
counter = 0 
for contour in contours :
    if cv2.contourArea(contour) > 10  :
        counter+=1
        x,y,w,h = cv2.boundingRect(contour)
        #print(contours)
        cv2.rectangle(res , (x,y) , (x+w , y+h) , (127,127,127) , 3) #bounding box 
        continue
print(counter)
plt.imshow(res , cmap="gray")
plt.show()
```
