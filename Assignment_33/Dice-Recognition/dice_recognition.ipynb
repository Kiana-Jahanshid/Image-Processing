```python
import cv2
import matplotlib.pyplot as plt 
import numpy
```

```python
image = cv2.imread("input/dice.png" , cv2.IMREAD_GRAYSCALE)
plt.imshow(image  , cmap="gray")
plt.axis("off")
plt.show()
```

```python
_ , thresh = cv2.threshold(image , 80 , 255  , cv2.THRESH_BINARY)
plt.imshow(thresh , cmap="gray")
plt.axis("off")
plt.show()
```

```python
contours , hierarchy = cv2.findContours(thresh , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE) # ax midim , liste contour ha ro mide 
print(hierarchy)
dice = image.copy()
count = 0
array = []
temp =  []
number_of_dices = 0 
for contour in contours :
    if cv2.contourArea(contour) > 100 and cv2.contourArea(contour) < 1000:
        x , y , w , h = cv2.boundingRect(contour)
        cv2.rectangle(dice , (x-4,y-4) , (x+w+4 , y+h+4) , (127 , 127 , 127) , 3)
        count+=1

for i in range(len(hierarchy[0])):
        array.append(hierarchy[0][i][3])


for i in range(len(array)):
     if array[i] == -1 :
          number_of_dices +=1
print("number of deces : " , number_of_dices)


  
```

```python
plt.axis("off")  
plt.text(image.shape[0] /12 , image.shape[0]/13 , f"NUMBER of DICE :{number_of_dices}" )
plt.text(image.shape[0] /12 , image.shape[0]/7 , f"Dice Number:{[count]}" )
plt.imshow(dice , cmap="gray") 
plt.show()   
```
