```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("SuperMan.jpg")
img = cv2.cvtColor(image , cv2.COLOR_BGR2HSV)

sky = cv2.imread("Assignment_36\inputs\sky.jpg")
sky = cv2.resize(sky, (image.shape[1], image.shape[0]))
```

```python
H , S , V = cv2.split(img)
H_new = H.copy().astype(np.float32)  

for i in range(H.shape[0]):
    for j in range(H.shape[1]):
        if 35 <H_new[i,j] < 85  and S[i,j] > 75 :
            image[i,j] = sky[i,j]


img_final = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.imshow(img_final)
```
