```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("cl.jpg")
img2 = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)

H , S , V = cv2.split(img)
H_new = H.copy().astype(np.float32)
```

```python
H_new2 = H.copy().astype(np.float32) 
for i in range(H.shape[0]):
    for j in range(H.shape[1]):

        if  S[i,j] > 150 : # for keeping skin unchanged

            if 90 <= H[i,j] <= 105 : # blue to  yellow
                H_new2[i,j] -= 75

            if 105 < H[i,j] <= 130 : # blue to  yellow
                H_new2[i,j] -= 80


            if 0 <= H[i,j] <= 5 : # red to  green
                H_new2[i,j] += 65
                
            if 170 <= H[i,j] <= 180 : # red to  green
                H_new2[i,j] += 70

H_new2 = H_new2.astype(np.uint8)
result = cv2.merge((H_new2,S,V))
result = cv2.cvtColor(result , cv2.COLOR_HSV2RGB)
plt.imshow(result)
```
