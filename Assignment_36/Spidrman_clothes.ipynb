```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("spiderman.jpg")
img2 = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
img = cv2.cvtColor(img , cv2.COLOR_BGR2HSV)

H , S , V = cv2.split(img)
H_new = H.copy().astype(np.float32) #refrence by pointer - pointer H ha yeki mishe , har ja H ro taghir bedim , H  asli ham taghir mikone 
```

```python
for i in range(H.shape[0]):
    for j in range(H.shape[1]): 

        if H_new[i,j] < 0 :      # green 
            H_new[i,j] += 187

        if H[i,j] <15 or H[i,j] > 165 :  
            H_new[i,j] += 60 

        if 90 < H[i,j] <= 100 :
            H_new[i,j] -= 155
            
        if 100 < H[i,j] <= 120 :
            H_new[i,j] -= 150

        if 120 < H[i,j] < 140 :
            H_new[i,j] -= 145

        if 140 <= H[i,j] < 150 :
            H_new[i,j] -= 135

H_new = H_new.astype(np.uint8)
result = cv2.merge((H_new,S,V))
result = cv2.cvtColor(result , cv2.COLOR_HSV2RGB)
plt.imshow(result)
```

```python
H_new2 = H.copy().astype(np.float32) 
for i in range(H.shape[0]):
    for j in range(H.shape[1]): 

        if 0 <= H[i,j] <= 20 : #yellow
            H_new2[i,j] += 28

        if 170 <= H[i,j] <= 180 : 
            H_new2[i,j] += 35


        if 90 < H[i,j] <= 100 :  # green 
            H_new2[i,j] -= 50

        if 100 < H[i,j] <= 120 :
            H_new2[i,j] -= 50

        if 120 < H[i,j] < 140 :
            H_new2[i,j] -= 45

        if 140 <= H[i,j] < 150 :
            H_new2[i,j] -= 40

H_new2 = H_new2.astype(np.uint8)
result = cv2.merge((H_new2,S,V))
result = cv2.cvtColor(result , cv2.COLOR_HSV2RGB)
plt.imshow(result)
```
