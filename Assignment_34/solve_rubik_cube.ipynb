```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Assignment_34/inputs/rubik.png', cv2.IMREAD_UNCHANGED)
all_channels = img[:,:,0:3]
```

```python
#########  solve_red_side   #########

#blue + green = cyan
red_intensity_min = (100,100, 0)
red_intensity_max = (255,255,20)
mask = cv2.inRange(all_channels, red_intensity_min , red_intensity_max )

value = all_channels.copy()
value[mask==255] = (0,0,255)

cv2.imwrite("red_partof_cube.png" , value)
red_out = cv2.imread("red_partof_cube.png")
```

```python
#########  solve_blue_side   #########

#red + green = yellow
blue_intensity_min = (0,100,100)
blue_intensity_max = (100,255,255)
mask2 = cv2.inRange(red_out, blue_intensity_min , blue_intensity_max )

value[mask2==255] = (255,0,0)

cv2.imwrite("blue_partof_cube.png" , value)
blue_out = cv2.imread("blue_partof_cube.png")
```

```python
#########  solve_green_side   #########

#red + blue = magenta
green_intensity_min = (100,0,100)
green_intensity_max = (255,30,255)
mask3 = cv2.inRange(blue_out, green_intensity_min, green_intensity_max)

value[mask3==255] = (0,255,0)

cv2.imwrite("green_partof_cube.png" , value)
green_out = cv2.imread("green_partof_cube.png")
```
