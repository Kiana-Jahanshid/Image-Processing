```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
bg = np.ones(shape=(500,800,3), dtype=np.int16) * 255
r = 400
rainbow_list_color = [(250,0,0) , (255,160,0) , (255,255,0) , (0,240,0) ,(0,255,255) , (0,0,245) , (150,0,150) , (255,255,255)]
for i in range (8) :
    cv2.circle(img= bg, center = (400,500), radius = r, color =rainbow_list_color[i], thickness=-1)
    r-=40
plt.imshow(bg)
```
