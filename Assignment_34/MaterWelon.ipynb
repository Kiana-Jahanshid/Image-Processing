```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
watermelon = cv2.imread("Assignment_34\inputs\watermelon.jpg")
watermelon = cv2.cvtColor(watermelon , cv2.COLOR_BGR2RGB)
plt.imshow(watermelon)
```

```python

R , G , B  = cv2.split(watermelon)
MATERWELON1 = cv2.merge((G,B,B))
cv2.imwrite("RED.jpg" , MATERWELON1)
plt.imshow(MATERWELON1)
```
