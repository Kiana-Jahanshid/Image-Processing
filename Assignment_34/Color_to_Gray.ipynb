```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
color = cv2.imread("Assignment_34\inputs\COLOR.jpg")
color = cv2.cvtColor(color , cv2.COLOR_BGR2RGB)
plt.imshow(color)
```

```python
def color_to_grayscale(R,G,B):
    Grayscale = ((R + G + B )/3)
    #OR Grayscale = ((R + G + B ))
    return Grayscale


R , G , B  = cv2.split(color)
Grayscale = color_to_grayscale(R,G,B)

cv2.imwrite("gray.jpg" , Grayscale)
plt.imshow(Grayscale , cmap="gray")
```
